```python
# Importing 
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
#Storing CSV files into Dataframe
Csv1 = "Retail_Sales_Data/Merged_Features_Sales.csv"
Csv2="Cleaned_Sales.csv"
```

```python
#Loading first data frame
df1 = pd.read_csv(Csv1)
df1.head()
```

```python
df1.shape
```

```python
#Loading Second data frame
df2 = pd.read_csv(Csv2)
df2.head()
```

```python
 df2.duplicated(['Sale_Date'] , keep = 'first')
df2.shape
```

```python
#Renaming df2 columns to match df1
df2.rename(inplace= True, columns= {'Store': 'store', 'Weekly_Sales':'weekly_sales','Sale_Date':'sale_date'})
df2
```

```python
#Dropping unnecesary columns 
df2.drop(inplace= True,columns=['IsHoliday', 'Dept',])
df2
```

```python
#new= pd.merge(df1, df2, left_on="weekly_sales")
new = pd.merge(df1, df2, how='inner', on='weekly_sales')
```
