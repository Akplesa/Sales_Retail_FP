```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import os
```

```python
df = pd.read_csv('Retail_Sales_Data/Final_Merged_FeatureSales.csv')
```

```python
df
```

```python
# Dropping the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Dropping the null rows
df = df.dropna()
df.head()
#Getting rid of space on FALSE_POSITIVES
df.replace(to_replace ="FALSE POSITIVE", 
                 value ="FALSE_POSITIVE",
          inplace=True)
```

```python
#Drop all other unnecesary columns 
df.drop(inplace= True,columns=['temperature', 'fuel_price', 'store','unemployment','isholiday','dept'])
```

```python
df
```

```python
#Sort Values by date
df.sort_values(by='sale_date', inplace= True) 
```

```python
#use conditional to locate all rows after November 4th 2011
new_df= df.loc[df["sale_date"]> "2011-11-04" ]
```
