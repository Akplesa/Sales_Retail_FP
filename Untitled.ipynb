```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv("Retail_Sales_Data/Official_Merged_SF.csv").dropna()
```

```python
df
```

```python
# assigning the data to X and y
X = df[["markdown1", "markdown2", "markdown3", "markdown4", "markdown5"]]
y = df["weekly_sales"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
# assigning the data to X and y
X = df["isholiday"].values.reshape(-1,1)
y = df["markdown5"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# #  Create the model using LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
